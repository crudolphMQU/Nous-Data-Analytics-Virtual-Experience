# Seaborn plotting


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(data = df, x = 'year', y = 'student_teacher_ratio', hue = 'state_territory')
```

All three states show an increase from the years 2012 to 2016 followed by a decrease from 2016 to 2018. NSW has a sharp increase 2018 onwards. The other states have further decreases from 2018 onwards. We also notice from the period of 2013 - 2018, all states had similar ratios. 

What could have happened in 2018 that caused this sharp increase in NSW? 

Could COVID have affected the ratio from 2019 onwards? 

```python
ax = sns.lineplot(x = 'year', y = 'student_teacher_ratio', 
                  hue = 'state_territory',
                  data = df)

ax.set(xlabel='Year', ylabel='Student to teacher ratio')
legend_handles, _= ax.get_legend_handles_labels()
ax.legend(legend_handles, ['NSW','VIC','QLD'], bbox_to_anchor=(1,1))
plt.title('Student-Teacher Ratio For Government-Run Secondary Schools')
plt.show()
```
